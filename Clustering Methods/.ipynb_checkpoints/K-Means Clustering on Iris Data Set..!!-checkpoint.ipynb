```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
```

```python
from sklearn import datasets

#Iris DataSet
iris = datasets.load_iris()
X = iris.data
```

```python
#K-Means
kmn = KMeans(n_clusters = 3)
kmn.fit(X)
labels = kmn.labels_
```

```python
#Plotting
fig = plt.figure(1,figsize=(9,9))
ax = Axes3D(fig,rect=[0,0,0.95,1],elev = 48,azim=134)
ax.scatter(X[:,3],X[:,0],X[:,2], c = labels.astype(np.float),edgecolor = "c",s=50)
ax.set_xlabel("Petal Width")
ax.set_ylabel("Sepal Length")
ax.set_zlabel("Petal Length")
plt.title("K-Means",fontsize = 18)
plt.show()
```

```python
#The one problem with KMeans clustering is that it does "Hard Labelling"....that is it tries to predict
#the class on the based of Maximum probability.
#To avoid this or overcome this...we try another Model"Mixture Models".
```
